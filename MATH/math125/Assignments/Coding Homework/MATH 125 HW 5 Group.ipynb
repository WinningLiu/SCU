```python
import pandas as pd
import numpy as np
from google.colab import files
import io
import matplotlib.pyplot as plt

uploaded1 = files.upload()
uploaded2 = files.upload()
df1 = pd.read_csv(io.BytesIO(uploaded1['30dayTreasury.csv']))
df2 = pd.read_csv(io.BytesIO(uploaded2['SP_VWRETD.csv']))

#Part a
print("Mean for 30 day treasury:", np.mean(df1['t30ret'][0:]))
print("Standard deviation for 30 day treasury:", np.std(df1['t30ret'][0:]))

print("Mean for S&P 500:", np.mean(df2['vwretd'][0:]))
print("Standard deviation for S&P 500:", np.std(df2['vwretd'][0:]))

#Part b
bonds = []
x1 = []
temp = 1
for i in range(0, len(df1['t30ret'])):
  temp = temp * (1+df1['t30ret'][i])
  bonds.append(temp)
  x1.append(i+1)

plt.title("Bonds returns")
plt.scatter(x1, bonds)
plt.show()
print("Bond price after 50 years:", bonds[len(bonds) - 1])

stocks = []
x2 = []
temp = 1
for i in range(0, len(df2['vwretd'])):
  temp = temp * (1+df2['vwretd'][i])
  stocks.append(temp)
  x2.append(i+1)

plt.title("S&P Returns")
plt.scatter(x2, stocks)
plt.show()
print("S&P Price after 50 years:", stocks[len(stocks) - 1])

#Part c
temp = 1
maximum = []
x3 = []
for i in range(0, len(df1['t30ret'])):
  temp = temp * (1 + max(df1['t30ret'][i], df2['vwretd'][i]))
  maximum.append(temp)
  x3.append(i + 1)

plt.title("Maximum Returns")
plt.scatter(x3, maximum)
plt.show()

print("Maximum return:", maximum[len(maximum) - 1])
```
