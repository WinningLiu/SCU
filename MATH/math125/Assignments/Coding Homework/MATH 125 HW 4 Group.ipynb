```python
from scipy.stats import norm
import numpy as np
import pandas as pd
from datetime import datetime
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import math
import random
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = np.array([1,2,3,4,5,6,7]).reshape((-1, 1))
y = np.array([0.00852, 0.01045, 0.00573, 0.01578, -0.01223, 0.00682, 0.00452])
model = LinearRegression().fit(x, y)
plt.scatter(x,y)
plt.plot(x, model.predict(x))
plt.show()

#Part a
a = model.intercept_
b = model.coef_[0]
print("a + bt =", a, "+", b, "t")

#Part b
#The intercept from part a and the average returns are close, but not the same
#They are close since if you are trying to minimize the differences, then the 
#value should be close to the mean
averageReturns = np.mean(y)
print("Constant a:", a)
print("Average returns:", averageReturns)

#Part c
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(x)
poly.fit(X_poly, y)
lin2 = LinearRegression().fit(X_poly, y)

plt.scatter(x, y)
plt.plot(x, lin2.predict(poly.fit_transform(x)))
plt.show()

a = lin2.intercept_
b = lin2.coef_[1]
c = lin2.coef_[2]

print("a + bt + ct^2 =", a, "+", b, "t +", c, "t^2")
```
