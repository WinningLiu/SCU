```python
import numpy as np
import pandas as pd
import math
from scipy import linalg
import matplotlib.pyplot as plt
from scipy.stats import norm

h  = (1/252)
k = 2*math.sqrt(h)
a = (1/2)*(h/k**2)
b = (1 - (h/k**2))

s = math.ceil(10/k)
f0 = np.ones(s) * (1/1000000)
Q = np.matrix([np.zeros(s)] * s)

Q[0,0] = b
Q[0,1] = a

for i in range(1, s-1):
  Q[i, i-1] = a
  Q[i, i] = b
  Q[i, i+1] = a

Q[s-1, s-2] = a
Q[s-1, s-1] = b

#f = Q * f0
f = np.linalg.matrix_power(Q, 252).dot(f0.reshape(-1, 1))
```

```python
plt.plot(f)
plt.show()
```
